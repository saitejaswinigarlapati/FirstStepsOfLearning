### Simple Calculator

```python
def add(numbers):
    return sum(numbers)

def sub(numbers):
    result=numbers[0]

    for num in numbers[1:]:
        result -=num
    return result

def mul(numberrs):
    result=1
    for num in numbers:
        result*=num
    return result

def div(numbers):
    result=numbers[0]
    for num in numbers:
        if num==0:
            raise ZeroDivisionError("Can not divide by zero")
        result /=num
    return result

        

def main():
    print("Simple Calculator")
    print("Select operator")
    print("1. Add")
    print("2.Substraction")
    print("3. Multiplicaation")
    print("4.Division")

    while True:
        choice=input("Enter choice 1/2/3/4 : ")

        if choice in ("1","2","3","4"):
            try :
                numbers=list(map(float,input("Enter numbers with space seperation : ").split()))
                if len(numbers)<2:
                    print("Enter atleast 2 numbers")
                    continue
            except ValueError:
                print("Invalid innput .please enter a number .")
                continue
        if choice=="1":
            print(f"result ={add(numbers)}")

        elif choice=="2":
            print(f"result = {sub(numbers)}")
        elif choice=="3":
            print(f"result ={mul(numbers)}")
        elif choice=="4":
            try:
                print(f"result ={div(numbers)}")
            except ZeroDivisionError:
                print("can not divide with zero")
        else:
            print("Invalid input . please enter a valid choice 1/2/3/4")
        next_calc=input("Do you want to perform another calculation ? (yes/no) : ")
        if next_calc.lower() != "yes":
            break
print("started...")
main()
print("end..")
```

```python
def calculate(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero"
    except Exception:
        return "Error: Invalid expression"

def main():
    print("Welcome to the Real Calculator!")
    print("Type your expression like 12 + 3 * 2 and hit ENTER.")
    print("Type '=' at any time to get the result.")
    print("Type 'exit' to quit.\n")

    expression = ""
    while True:
        user_input = input(">>> ")

        if user_input.strip().lower() == "exit":
            print("Goodbye!")
            break
        elif user_input.strip() == "=":
            result = calculate(expression)
            print("= ", result)
            expression = str(result) if isinstance(result, (int, float)) else ""
        else:
            expression += user_input


main()
```

Rock Paper Scissors Game

```python
import random
def choice():
        player=input("Enter your choice (rock/paper/scissors) : ")
        options=["rock","paper","scissors"]
        computer=random.choice(options)
        return player , computer

def check_win(player,computer):
        if player==computer:
                return "It's a tie!"
        elif player=="rock":
                if computer=="paper":
                        return "You lose!"
                else:
                        print("You win!")
        elif player=="paper":
                if computer=="rock":
                        return "You win."
                else:
                        return "You lose!"
        elif player=="scissors":
                if computer=="paper":
                        return "You win."
                else:
                        return "You lose!"
        
                
while True:
        player,computer=choice()
        player,computer=player.lower(),computer.lower()
        print(f"You choice :{player.capitalize()}\nComputer choice :{computer.capitalize()}")
        result=check_win(player,computer)
        print(result)
        a=input("Do you want to play again ? (yes/no) ")
        if a.lower()=="yes":
                continue
        else:
                break
```
