```python
# numbber guessing game

import random

n=random.randrange(21)

print("Welcome to Number Guessing Game!")
g=int(input("enter a number betweem 0 and 20 : "))
c=1
if g>20:
    print("please enter a number between 0 to 20")
for i in range(5):
    if c<5:
    
        if n<g:
            print("Too high")
            print(f"You hove {5-c} attempts")

            g=int(input("guess again : "))
        elif n>g:
            print("Too low")
            print(f"You have only {5-c} attempts")

            g=int(input('guess again : '))
        c+=1

if g==n:
    print("Yes,your guess is correct ")
else:
    print("U loss! Better luck next time.")
    print("correct guess is : ",n)
   
```

```python
#checking the given string are anagrans or not

s1=input("Enter a string : ")
s2=input("Enter a string : ")
l=[]
c=0
if len(s1)==len(s2):
    for i in s1:
        for j in s2:
            if i==j and i not in l:
                l.append(i)
                c+=1
    if c==len(s1):
        print("Given Strings are Anargams")
    else:
        print("Given Strings are not Anagrams")
else:
    print("Given Strings are not Anagrams")
```
