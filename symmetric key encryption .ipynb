```python
#importing Fernet class from the cryptography module
from cryptography.fernet import Fernet

#generating a key
key=Fernet.generate_key()

#creating a instant of Fernet class with key
fernetkey=Fernet(key)

#our message
m="HI ,HOW R U?"

#encrypting the message
em=fernetkey.encrypt(m.encode())
print("em : ",em.decode())

#decrypting the message
dm=fernetkey.decrypt(em.decode())
print("dm: ",dm.decode())
```
