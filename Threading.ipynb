```python
#importing threading and time modules
import threading
import time

def fun():
    for i in range(5):
        print("HELLO")
        time.sleep(1)
def run():
    for i in range(5):
        print("HIII")
        time.sleep(1)
        
# creating "Thread" class  from the tradeing module
t1=threading.Thread(target=fun)
t2=threading.Thread(target=run)

#starting the thread's
t1.start()
#creating small delay b/w t1 and t2 to avoid collision
time.sleep(0.2)
t2.start()

#telling main thread to wait untill t1,t2 excutes completely
t1.join()
t2.join()

print("BYE")
```
