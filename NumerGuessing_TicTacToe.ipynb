### NUMBER GUESSING GAME

```python
import random
import time

n=random.randrange(1,101)

print('Welcome to the NUMBER GUESSING GAME \n u have only 5 attemps to guess the number')

time.sleep(2)

g=int(input('guess number : '))

C=1

while n!=g and C<5:
    if g<n:
        print('too low')
        print(f"u have {5-C} attemps left")
        g=int(input('guess again : '))
        C+=1
    elif g>n:
        print('too high')
        print(f"u have {5-C} attemps left")
        g=int(input('guess again : '))
        C+=1
    else:
        break
if n==g:
    print("*** u r guess is correct ***")  
else:
    print('****better luck next time****')
    print(f"Correct guess : {n}")
```

### Tic-Tac-Toe Game

```python
board={'1':' ','2':' ','3':' ',
       '4':' ','5':' ','6':' ',
       '7':' ','8':' ','9':' '}
def printboard(board):
    print(board['7']+'|'+board['8']+'|'+board['9'])
    print("-+-+-")
    print(board['4']+'|'+board['5']+'|'+board['6'])
    print("-+-+-")
    print(board['1']+'|'+board['2']+'|'+board['3'])
def game():
    turn='x'
    c=1
    for i in range(9):
        printboard(board)
        
        print("it's ur turn "+ turn + ". choose u r place")
        
        move=input()
        
        print('c',c)
        
        if board[move]==' ':
            board[move]=turn
            c+=1
        else:
            print('that palce is alredy filled . \n choose another place')
            continue
        if c>=4:
            if board['1']==board['5']==board['9']!=' ':
                printboard(board)
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break
         
    
       
            if board['2']==board['5']==board['8']!=' ':
                printboard(board) 
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break

            if board['3']==board['6']==board['9']!=' ':
                printboard(board) 
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break
            if board['7']==board['8']==board['9']!=' ':
                printboard(board) 
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break
            if board['4']==board['5']==board['6']!=' ':
                printboard(board) 
                print('\n game over.\n')
               
                break
            if board['4']==board['1']==board['7']!=' ':
                printboard(board) 
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break
            
            if board['1']==board['2']==board['3']!=' ':
                printboard(board) 
                print('\n game over.\n')
                print(f' GAME OVER {turn} won the game')
                break
                
        elif c==9:
            print('game over')
            print('it\'s a tie')
            break
        if turn=='x':
            turn='o'
        else:
            turn='x'
game()
```
