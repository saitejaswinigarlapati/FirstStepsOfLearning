### Syntax Error

```python
# map
def add(a,b):
    return a+b
l=[1,2,3,4,5]
a=list(map(add(([x,y for x ,y in l]),l))
print(a)
```

```python
def add(a, b):
    return a + b

l = [1, 2, 3, 4, 5]
b = 4
a = list(map(lambda x: add(x, b), l))
print(a)  # Output: [5, 6, 7, 8, 9]
```

```python
#reduce 

from functools import reduce

expen=[("Dinner", 80),
      ("car",100)]

sum=reduce(lambda a,b :a[1]+b[1],expen)
print(sum)

```

```python
def to_upper(fun):
    def wrap():
        print("Before")
        message=fun()
        upper=message.upper()
        print("After")
        return upper
    return wrap

@to_upper
def great():
    return "Hello"
```
