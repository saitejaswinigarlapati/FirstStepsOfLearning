```python
def print_pattern(n):
    # Upper part
    for i in range(n):
        print("  " * (n - i), end="")  # Print spaces
        print("* " * (i + 1))  # Print stars

    # Lower part
    for i in range(1, n):
        print("  " * (i + 1), end="")  # Print spaces
        print("* " * (n - i))  # Print stars

# Example usage
n = 5
print_pattern(n)
```
