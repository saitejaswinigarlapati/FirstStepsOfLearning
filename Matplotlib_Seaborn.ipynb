```python
#subplot's

plt.subplot(2,2,1) #dimensions of graph
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,c='r')
plt.subplot(2,2,4)
plt.plot(x,y,c='g')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave')
plt.plot(x,y)
```

```python
#Barplot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Histogram
a=np.array([1,11,20,51,5,79,31,27,22,87,55,43,56,73,54])
plt.hist(a)
plt.show()
```

```python
#Box Plot using matplotlib
#helps to find percentile
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [715, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue' ]
explode = (0.1, 0, 0, 0) # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
