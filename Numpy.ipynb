```python
array1=np.arange(0,10)
array2=array1
print(array1)
array2[3:]=23
print("array1 : ",array1)
print("array2 : ",array2)
#arrays are reference type .
# if we change value in array1 it also changes in array2
#so we use copy function .
```

```python
array1=np.arange(1,5)
array2=array1.copy()
array1[2:4]=9
print("array 1 :",array1)
print("array 2 :",array2)
```
