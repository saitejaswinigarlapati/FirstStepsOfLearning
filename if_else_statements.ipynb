```python
# MAX OF 3
x,y,z=map(int,input("Enter 3 numbers : ").split())
if x>y and x>z:
    print("max of x,y,z is",x)
elif y>x and y>z:
    print("max of x,y,z is",y)
elif z>x and z>y:
    print("max of x,y,z is",z)    
```

```python
#GRADE
m=int(input())
if m<25:
    print("F")
elif m>=25 and m<45:
    print("E")
elif m>=45 and m<50:
    print("D")
elif m>=50 and m<60:
    print("C")
elif m>=60 and m<80:
    print("B")
elif m<80:
    print("A")
```
