## Pandas is the popular library offen used in data analysis


```python
import pandas as pd
import numpy as np
data=np.arange(1,21).reshape(5,4)
print("data : \n")
print(data)
```

```python
print("DataFrame") 
a=pd.DataFrame(data,index=['a','b','c','d','e'],columns=['x','y','z','w']) # ceating a datafrrame 
print(a)
```

```python
print("Coverting DataFrame into a csv file : ")
b=a.to_csv ('csv_dataframe') # creates a file 
print("csv file created")
```

```python
print("Accessing the elements using .loc  : ")
print(a.loc['a']) # .loc uses row names
print("type :",type(a.loc['a']))
```

```python
print("Acessing the elements using .iloc : ")
print(a.iloc[0:2,0:2]) # .iloc usies only indexing 
print(type(a.iloc[0:2,0:2]))
```

```python
print(a.loc[['a','b']]) # for multiple row we use list 
print("Data type :")
print(type(a.loc[['a','b']]))
```

```python
# Converting DaataFrame to array 
array=a.iloc[:,:].values   # .values --> converts data into array
print(array)
```

```python
print("null values : ")
print(a.isnull()) # checking for the null values 
print("No.of times each value repeated in the row :")
print(a['x'].value_counts()) # counts the each element 
print("UNique values in a row :")
print(a['x'].unique())
```
